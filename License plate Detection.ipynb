## License Plate Detection using OpenCV and Haar Cascade Classifier
**Name:** YUVAN M **Reg No:** 212223240188

**Slot:** 4E6-1

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('input.jpg',1)
gray = cv2.cvtColor(img, 0)
```

```python
cascade_path = cv2.data.haarcascades + 'haarcascade_russian_plate_number.xml'
```

```python
plate_cascade = cv2.CascadeClassifier(cascade_path)
plates = plate_cascade.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=4,minSize=(30, 30))
print(f"Total license plates detected: {len(plates)}")
```

```python
for i, (x, y, w, h) in enumerate(plates):
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
    plate_roi = img[y:y + h, x:x + w]
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10, 6))
plt.imshow(img_rgb)
plt.axis('off')
plt.title('Detected License Plates')
plt.show()
```
